```python
class Human():
    def __init__(self,name, age, gender, bg):
        self.name = name
        self.age = age
        self.gender = gender
        self.blood_gp =bg          
       
    def walk(self):
        print(f"Our human {self.name} is walking")
        
    def eat(self):
        print(f"Our human {self.name}is eating")
    
```

```python
human1 = Human()
```

```python
def fn_grading(percentage):
    grade = None
    
    if percentage>=80:
        grade = "A+"
    elif percentage>=70:
        grade =  "A"
    elif percentage>=60:
        grade =  "B"
    elif percentage>=50:
        grade = "C"
    elif percentage>=40:
        grade = "D"
    else:
        grade =  "F"
    return grade
```

```python
df['Grade'] = df['Percentage'].apply(fn_grading)
```
